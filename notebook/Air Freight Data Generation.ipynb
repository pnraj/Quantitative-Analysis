### Route Table

```python


# Define a list of countries
countries = ['United States', 'China', 'India', 'Brazil', 'Russia']
modes = ['Sea','Air']

# Create a dictionary to store distances between pairs of countries
country_distances = {}

# Populate the dictionary with approximate distances (in kilometers) between pairs of countries
modes_country = {}
for md in modes:
    for country1 in countries:
        for country2 in countries:
            if country1 != country2:
                # Generate a random distance (for demonstration purposes)
                distance = np.random.randint(500, 5000)
                # Add the distance to the dictionary for both orders of the countries
                country_distances[(country1, country2)] = distance
                country_distances[(country2, country1)] = distance  # Ensuring consistency
    modes_country[md] = country_distances

    
data = []
for mode, distances_dict in modes_country.items():
    for (origin, destination), distance in distances_dict.items():
        data.append([origin, destination, mode, distance])

dfr = pd.DataFrame(data, columns=['Origin', 'Destination', 'Mode of Transportation', 'Distance btw'])

# Display the DataFrame
dfr['Route ID'] = np.arange(1, len(dfr) + 1)
# Save DataFrame to a CSV file
dfr.to_csv('Routes_Table.csv', index=False)
```

### Service Table

```python
company_name = {"com1": 1, "com2": 2, "com3": 3, "com4": 4, "com5": 5, "com6": 6, "com7": 7, "com8": 8,"com9": 9, "com10": 10 }

rounte_id = np.arange(1, 41)

Serv_cost = {"First_Class": 0.15, "Business Class": 0.10, "Economy Class": 0.8}

Serv_table = []

for rid in rounte_id:
    for ids,com in company_name.items():
        for srv, cost in Serv_cost.items():
            Serv_table.append([rid,ids,com,srv, cost ])

ser_df = pd.DataFrame(Serv_table, columns=['Route ID','Company Name','Company ID','Service Level','Cost per km ($)'])
    
ser_df.to_csv('Service_Table.csv', index=False)       
```

### trip table

```python
import pandas as pd
import numpy as np
import datetime as dt

# Define the number of routes and companies
num_routes = 40
num_companies = 10

# Generate Route ID and Company ID
route_id = np.arange(1, num_routes + 1)
company_id = np.random.choice(np.arange(1, num_companies + 1), size=num_routes)

depart_dates = np.array([dt.datetime(2024, 2, 1, 8, 0) + dt.timedelta(minutes=np.random.randint(0, 480)) for _ in range(num_routes)])
estimated_times = np.array([np.random.randint(120, 301) for _ in range(num_routes)])  # 2 to 5 hours in minutes

# Convert depart_dates to datetime64 for compatibility
depart_dates = np.array(depart_dates, dtype='datetime64')

reach_timedeltas = np.array([np.timedelta64(int(est_time), 'm') for est_time in estimated_times])
# Generate reach dates
reach_dates = depart_dates + reach_timedeltas

# Generate delays
delays = np.array([np.random.randint(-60, 61) for _ in range(num_routes)])  # -1 to +1 hours in minutes
# Convert delays to Python integers
delays = delays.astype(int)

## Apply delays to reach dates
reach_dates += np.array(delays, dtype='timedelta64[m]')


# Create DataFrame for the trip table
trip_data = {
    'Route ID': route_id,
    'Company ID': company_id,
    'Depart DateTime': depart_dates,
    'Reach DateTime': reach_dates,
    'Estimated Time': estimated_times,
    'Delays': delays
}

trip_df = pd.DataFrame(trip_data)

# Display the DataFrame
trip_df.to_csv('Trip_Table.csv', index=False)
```
